# 12. 이미지 변형(원근)

원근을 쓰는 이유?
삐뚤빼뚤한 이미지나 사다리꼴 형태의 고르지 못한 이미지를   
우리가 원하는 직사각형 형태로 만들어주기 위해서
실제 많이 쓰이는 것은 원근 이미지를 평면 이미지로 변환하는 것.

## 사다리꼴 이미지 펼치기   


```python
import cv2
import numpy as np

img = cv2.imread('newspaper.jpg')

width, height = 640, 240 #가로 크기 640, 세로 크기 240으로 결과물 출력


src = np.array([[511,352], [1008,345], [1122,584], [455,594]], dtype=np.float32) #input
#src는 input을 지정해줌. input -> 사다리꼴 모양의 모서리 4개

dst = np.array([[0,0], [width, 0], [width, height], [0,height]], dtype=np.float32) #output
# dst = 좌상, 우상, 우하, 좌하(시계 방향으로 4지점 정의)

matrix = cv2.getPerspectiveTransform(src,dst) #Matrix 얻어옴
#getPerspectiveTransform 이란? src영역을 dst영역으로 바꾸기 위한 변환행렬을 얻어옴
result = cv2.warpPerspective(img, matrix, (width,height)) 
#warpPerspective 는? 변환행렬 값을 저장해서 최종결과 이미지를 얻어옴

cv2.imshow('img', img)
cv2.imshow('result', result)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## 회전된 이미지 올바로 세우기

```python
import cv2
import numpy as np

img = cv2.imread('poker.jpg')

width, height = 530, 710

src = np.array([[702,143], [1133,414], [726,1007], [276,700]], dtype=np.float32) #input
#기울어진 포커 카드위 위치(초록색 지점)

dst = np.array([[0,0], [width, 0], [width, height], [0,height]], dtype=np.float32)#output
# dst = 좌상, 우상, 우하, 좌하(시계 방향으로 4지점 정의)

matrix = cv2.getPerspectiveTransform(src,dst) #Matrix 얻어옴
result = cv2.warpPerspective(img, matrix, (width,height)) 

cv2.imshow('img', img)
cv2.imshow('result', result)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
