# 16. 이미지 변환(침식)

## 이미지를 깎아서 노이즈 제거

흰색 영역의 외곽 픽셀을 검은색으로 변경하기

## 원리

토양의 침식과 비슷하다.   
이미지의 경계부분을 침식시켜 배경이미지로 전환하는 것이 원리이다.   



```python
import cv2
import numpy as np

kernel = np.ones((3,3), dtype=np.uint8) #np.zeros는 공간을 전부 0.
#ones는 전부 1로 채움
#kernel
 
img=cv2.imread('erode.png', cv2.IMREAD_GRAYSCALE) #이미지 불러오기
erode_1=cv2.erode(img, kernel, iterations=1)#반복횟수
erode_2=cv2.erode(img, kernel, iterations=2)
erode_3=cv2.erode(img, kernel, iterations=3)

cv2.imshow('origin img', img)

#검은 픽셀로 점점 크게 채워나감
cv2.imshow('erode_1', erode_1)
cv2.imshow('erode_2', erode_2)
cv2.imshow('erode_3', erode_3)


cv2.waitKey(0)
cv2.destroyAllWindows()
```
