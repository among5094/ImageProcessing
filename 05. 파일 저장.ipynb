## 동영상 저장

```python
import cv2
cap = cv2.VideoCapture('video_earth.mp4')

#코덱 정의(어떤 형태로 저장할지) cc-> 캐릭터코드, 한 글자씩 따로 떨어뜨려주는 역할 (*'DIVX')->  'D','I','V','X' 이렇게
fourcc = cv2.VideoWriter_fourcc(*'DIVX')

#프레임의 가로 세로 크기와 재생 속도(fps)정의
#width에는 정수값이 들어가야 하므로 round라는 함수로 감싸기->혹시 실수가 들어갈 수도 있으니까
width = round(cap.get(cv2.CAP_PROP_FRAME_WIDTH)) #영상의 width값을 가지고 옴
height = round(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv2.CAP_PROP_FPS)*2 #영상재생속도

out = cv2.VideoWriter('output_earth.avi', fourcc, fps, (width, height) ) #out객체생성
#(저장 파일 명, 코덱, 재생속도(fps), 크기(width, height))

while cap.isOpened(): #캡이 잘 열려있으면?
    ret, frame = cap.read() #성공하면 cap.read()를 frame에 저장
    
    if not ret:
        break
    out.write(frame)#영상데이터만 저장, 소리는 x
    cv2.imshow('video', frame)
    if cv2.waitKey(1) == ord('q'):
        break
            
out.release() #out객체 해제             
cap.release()
cv2.destroyAllWindows()
```
